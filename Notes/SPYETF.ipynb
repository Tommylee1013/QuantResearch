### SPY Index

SP500 index ETF

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

start_date = '2014-01-01'
end_date = '2024-03-13'

data = yf.download('SPY', start = '2014-01-01', end = '2024-03-13')

data['Returns'] = data['Close'].pct_change()
data['Cumulative Returns'] = (1 + data['Returns']).cumprod() - 1

returns_after_5_days = []
for i in range(len(data)):
    for j in range(i, len(data)):
        if data.iloc[j]['Cumulative Returns'] - data.iloc[i]['Cumulative Returns'] <= -0.03:
            target_day = j + 5 
            if target_day < len(data):
                returns_after_5_days.append(data.iloc[target_day]['Cumulative Returns'] - data.iloc[j]['Cumulative Returns'])
            break
```

```python
plt.hist(returns_after_5_days, bins=50, alpha=0.75, edgecolor = 'none')
plt.xlabel('Returns after 5 days from -3% Cumulative Return')
plt.ylabel('Frequency')
plt.title('Distribution of Returns | 5 Days | -3% Cum.Return')
plt.grid(False)
plt.show()
```

#### 날짜를 기준으로 했을 때 결과

```python
def getStats(ret_list) :
    mean = np.array(
        ret_list
    ).mean()
    std = np.array(
        ret_list
    ).std()
    skew = pd.Series(
        ret_list
    ).skew()
    kurt = pd.Series(
        ret_list
    ).kurt()
    return [mean, std, skew, kurt]
```

```python
def getStatFrame(day_list) :
    day_df = pd.DataFrame()
    for i in day_list :
        day_df = pd.concat(
            [day_df, pd.Series(getStats(i))],
            axis = 1
        )
    day_df.index = ['mean','std','skew','kurt']
    day_df.columns = ['3%','5%','7%','9%','12%','15%','20%']
    day_df = day_df.T
    
    return day_df
```

```python
def getStatFrame_percent(day_list) :
    day_df = pd.DataFrame()
    for i in day_list :
        day_df = pd.concat(
            [day_df, pd.Series(getStats(i))],
            axis = 1
        )
    day_df.index = ['mean','std','skew','kurt']
    day_df.columns = ['5 days','7 days','9 days','11 days','13 days','15 days','20 days','30 days','60 days']
    day_df = day_df.T
    
    return day_df
```
