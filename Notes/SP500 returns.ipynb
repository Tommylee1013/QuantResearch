### S&P500 Returns

S&P500 index에 대한 수익률

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = yf.download('^GSPC', start = '2014-01-01', end = '2024-03-07')

data['Returns'] = data['Close'].pct_change()
data['Cumulative Returns'] = (1 + data['Returns']).cumprod() - 1

returns_after_5_days = []
for i in range(len(data)):
    for j in range(i, len(data)):
        if data.iloc[j]['Cumulative Returns'] - data.iloc[i]['Cumulative Returns'] <= -0.03:
            target_day = j + 5 
            if target_day < len(data):
                returns_after_5_days.append(data.iloc[target_day]['Cumulative Returns'] - data.iloc[j]['Cumulative Returns'])
            break
```

```python
plt.hist(returns_after_5_days, bins=50, alpha=0.75)
plt.xlabel('Returns after 5 days from -3% Cumulative Return')
plt.ylabel('Frequency')
plt.title('Distribution of Returns | 5 Days | -3% Cum.Return')
plt.grid(False)
plt.show()
```
